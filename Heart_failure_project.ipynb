Step 1: Import Libraries & Load Dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv("heart.csv")
```

Step 3: Check Missing Values

```python
print(df.isnull().sum())
```

Step 4: Handle Invalid Values

```python
df['RestingBP'] = df['RestingBP'].replace(0, np.nan)
df['Cholesterol'] = df['Cholesterol'].replace(0, np.nan)
```

Step 5: Fill Missing Values

```python
df['RestingBP'].fillna(df['RestingBP'].median(), inplace=True)
df['Cholesterol'].fillna(df['Cholesterol'].median(), inplace=True)
```

Step 6: Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
print(df.shape)
```

Step 7 : Rename Column

```python
df.rename(columns={
    'Sex': 'Gender',
    'FastingBS': 'FastingBloodSugar',
    'RestingBP': 'RestingBloodPressure'
}, inplace=True)
```

Step 7: Clean Categorical Data

```python
df['Gender'] = df['Gender'].str.strip().str.upper()
df['ChestPainType'] = df['ChestPainType'].str.strip().str.upper()
df['RestingECG'] = df['RestingECG'].str.strip().str.upper()
df['ExerciseAngina'] = df['ExerciseAngina'].str.strip().str.upper()
df['ST_Slope'] = df['ST_Slope'].str.strip().str.upper()
```

Step 8: Fix Data Types

```python
df['Age'] = df['Age'].astype(int)
df['MaxHR'] = df['MaxHR'].astype(int)
df['HeartDisease'] = df['HeartDisease'].astype(int)

df['RestingBloodPressure'] = df['RestingBloodPressure'].astype(float)
df['Cholesterol'] = df['Cholesterol'].astype(float)
df['Oldpeak'] = df['Oldpeak'].astype(float)
```

Step 9 : Feature Engineering (New Columns)

```python
def age_group(age):
    if age < 40:
        return "Young"
    elif age < 60:
        return "Middle"
    else:
        return "Old"

df['AgeGroup'] = df['Age'].apply(age_group)
```

```python
def bp_category(bp):
    if bp < 90:
        return "Low"
    elif bp <= 120:
        return "Normal"
    else:
        return "High"

df['BP_Category'] = df['RestingBloodPressure'].apply(bp_category)
```
